```python
#string formatting method:there are three types of formating method in the string
#use format()method in the print()
#use format() in the end of the string
#use f infront of the string
```

```python
age = 20
text = "i am __ year old"
print (text.replace('__',str(age)))
```

```python
#using format()method in the print()
age = 20
text = "i am {} year old"
print (text.format(age))
```

```python
#using format at the end of the string
age = 20
text = "i am {} years old ".format(age)
print(text)
```

```python
#using f infront of the string
age = 20
text = f"i am {age} year old"
print(text)
```

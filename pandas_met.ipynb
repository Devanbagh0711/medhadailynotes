**combining and merging the dataframes**

````python
import pandas as pd
````

````python
#joining the data frames
d1 = pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6)}
)
d2 = pd.DataFrame(
    {"key":['b','c','e','f'],
     'num2':range(4)}
)
````

````python
print(d1)
print(d2)
````

````python
#merge the d1 and d2 where print only matching rows---inner join
pd.merge(d1,d2) #default pd.merge print the inner join
````

````python
#left join:- all the info from first table and matched info from 2nd table
pd.merge(d1,d2,how='left',on="key")
````

````python
#right join:-
pd.merge(d1,d2,how="right",on="key")
````

````python
#outer join:- merge all the data from table
pd.merge(d1,d2,how="outer",on="key")
````
